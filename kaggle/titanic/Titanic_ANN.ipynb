# Overfitting a Neural Net
In this notebook I will use the Titanic dataset from Kaggle to show an example of overfitting an artificial neural network. The model will then be significantly reduced but will be a much more robust net.

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')
```

```python
from google.colab import drive
drive.mount('/content/drive/', force_remount=True)
```

## Importing the dataset

```python
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/train.csv')
dummies = pd.get_dummies(data.Sex)
data = pd.concat([data,dummies], axis = 'columns')
dummies2 = pd.get_dummies(data.Embarked)
data = pd.concat([data,dummies2], axis = 'columns')
data.drop(['PassengerId','Name','Ticket','Cabin','Sex','male','Embarked','S'], axis = 1, inplace = True)
x = data.iloc[:,1:9].values
y = data.iloc[:,0].values
```

```python
data2 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/test.csv')
dummies = pd.get_dummies(data2.Sex)
data2 = pd.concat([data2,dummies], axis = 'columns')
dummies2 = pd.get_dummies(data2.Embarked)
data2 = pd.concat([data2,dummies2], axis = 'columns')
data2.drop(['PassengerId','Name','Ticket','Cabin','Sex','male','Embarked','S'], axis = 1, inplace = True)
x_test = data2.iloc[:,:].values
```

## Missing Data

```python
impute = SimpleImputer(missing_values = np.nan, strategy = 'mean')
impute.fit(x[:,1:3])
x[:, 1:3] = impute.transform(x[:, 1:3])
impute.fit(x_test[:,0:3])
x_test[:,0:3] = impute.transform(x_test[:, 0:3])
impute.fit(x_test[:,3:5])
x_test[:,3:5] = impute.transform(x_test[:, 3:5])
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x_test = sc.transform(x_test)
```

## Building and training the ANN

```python
ann = tf.keras.models.Sequential()

ann.add(tf.keras.layers.Dense(units=100, activation='relu', input_dim = 8))

ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))

ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

ann.fit(x, y, batch_size = 32, epochs = 500)
```

The accuracy of this model looks really good, over 92%, but we will see that this is actually an overfit net.

## Splitting into training and test sets to understand the overfitting problem

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

Traing the model on the split data:

```python
ann = tf.keras.models.Sequential()

ann.add(tf.keras.layers.Dense(units=100, activation='relu', input_dim = 8))

ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))

ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

ann.fit(X_train, y_train, batch_size = 32, epochs = 500)
```

Predicting on the split data:

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred = pd.DataFrame(y_pred)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y_pred.values) 
```

```python
accuracy = sum((y == y_test))/y.shape[0]
accuracy
```

### A smaller model

```python
ann2 = tf.keras.models.Sequential()

ann2.add(tf.keras.layers.Dense(units=10, activation='relu', input_dim = 8))

ann2.add(tf.keras.layers.Dense(units=10, activation='relu'))

ann2.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

ann2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

#ann2.fit(x, y, batch_size = 32, epochs = 500)
```

Predicting on the split data:

```python
y_pred2 = ann2.predict(X_test)
y_pred2 = (y_pred2 > 0.5)
y_pred2 = pd.DataFrame(y_pred2)
```

```python
y2 = le.fit_transform(y_pred2.values) 
accuracy2 = sum((y2 == y_test))/y2.shape[0]
accuracy2
```

The much smaller model has an accuracy of 86.76% on the training data while the large model has an accuracy of 92.13%, seemingly a large difference. However when we use the two models to make predictions on a new set of data the smaller model is at 86.59% accuracy and the large is 81.56%. Clearly the smaller model extrapoltes better to new data.
